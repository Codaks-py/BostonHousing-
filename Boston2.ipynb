```python
def corre(X_train, thre):
    corr_cols = set()
    corr = X_train.corr()
    for i in range(len(corr.columns)):
        for j in range(i):
            if abs(corr.iloc[i,j]) > thre:
                cols = corr.columns[i]
                corr_cols.add(cols)
                
    return corr_cols

corre(X_train, 0.75)
```

```python
def corre(X_test, thre):
    corr_cols = set()
    corr = X_test.corr()
    for i in range(len(corr.columns)):
        for j in range(i):
            if abs(corr.iloc[i,j]) > thre:
                cols = corr.columns[i]
                corr_cols.add(cols)
                
    return corr_cols

corre(X_test, 0.78)
```
